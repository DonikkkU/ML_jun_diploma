## Model Feature Importances

```python
with open('event_pred.pkl', 'rb') as f:
    rf_model = joblib.load(f)
```

```python

class Pipeline_setup:
    def __init__(self):
        self.df_final = pd.read_pickle('final_data')

    def pipeline(self):
        X = self.df_final.drop('target', axis=1)
        y = self.df_final['target']

        numerical_columns = X.select_dtypes(include=['int64', 'float64', 'int32']).columns
        categorical_columns = X.select_dtypes(include=['object']).columns
        categorical_columns = categorical_columns[~categorical_columns.isin(['session_id', 'client_id'])]

        numerical_transformer = Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ])

        categorical_transformer = Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('encoder', OneHotEncoder(handle_unknown='ignore'))
        ])

        preprocessor = ColumnTransformer(transformers=[
            ('numerical', numerical_transformer, numerical_columns),
            ('categorical', categorical_transformer, categorical_columns)
        ])

        rf_model = RandomForestClassifier(n_estimators=100, max_depth=None, random_state=42, min_samples_leaf=3, min_samples_split=2)

        pipe = Pipeline(steps=[
            ('preprocessor', preprocessor),
            ('classifier', rf_model)
        ])

        pipe.fit(X, y)

        # Получение важности признаков
        rf_model = pipe.named_steps['classifier']
        feature_importance = rf_model.feature_importances_
        for col in X.columns:
            # Рисование графика важности признаков
            plt.figure(figsize=(10, 6))
            plt.bar(range(len(feature_importance)), feature_importance)
            plt.xlabel('Признаки')
            plt.ylabel('Важность признака')
            plt.title('Важность признаков для RandomForestClassifier')
            plt.xticks(range(len(X[col])), X[col], rotation=45, fontsize=8) 
            plt.tight_layout()  
            plt.show()

        # Сохранение модели

#Pipeline
pipeline_setup = Pipeline_setup()
pipeline_setup.pipeline()
```
